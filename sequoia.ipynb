# Data cleaning 

### Temporarily fill NaN values with -1 for grouping

```python
apps = data
apps = apps.fillna(-1, inplace=True)
```

### Group duplicate productId's together 

```python
apps = data.groupby(['productId', 'categoryPath', 'startDate', 'activeUsers', 'avgActiveDays', 'totalMinutes', 'totalSessions'])
apps = apps.count().reset_index()
apps = apps.replace(-1, np.nan)
apps.head()
```

# Project out productId's growth rates using CMGR

### Get all unique apps, from first start date
recentUsers represents the number of monthly active users in the most recent month listed in the dataset for each unique mobile app
- Needed to calculate CMGR 

```python
unique = apps.copy()

recentUsers = unique.drop_duplicates('productId', keep='last')
recentUsers = recentUsers.activeUsers.reset_index().drop(columns=['index'])

# Keep the first instance of productId at earliest startDate
unique = unique.drop_duplicates('productId', keep='first') 
unique = unique.reset_index().drop(columns=['index'])
unique['recentUsers'] = recentUsers
unique.head()
```

```python
print('Dataset contains', len(unique), 'unique mobile apps')
```

### Calculate months active for each app
- Needed to calculate CMGR

```python
# Convert oldest startDate and pull date to type datetime
unique['startDate'] = pd.to_datetime(unique['startDate'], errors='coerce')
pullDate = pd.to_datetime('20200501', format='%Y%m%d', utc=True)

unique['monthsActive'] = (pullDate - unique['startDate']) / np.timedelta64(1, 'M')
unique['monthsActive'] = unique['monthsActive'].round(decimals=0)
unique.head()
```

### Calculate CMGR for each app and insert into new column
CMGR = (((EV/BV)^1/n) - 1) * 100

```python
unique['cmgr'] = (((unique.recentUsers / unique.activeUsers) ** (1 / unique.monthsActive)) - 1) * 100
unique.head()
```

```python
unique.plot(x='monthsActive', y='cmgr', kind='scatter')
```

## Remove outliers and implement cutoffs

```python
allGames = unique[unique.categoryPath.str.contains('game', case=False)]
print('Proportion of games to all mobile apps =', (len(allGames) / len(unique)))
```

### Remove games from dataset 
I removed all mobile apps under any 'game' category for a few reasons:
1. One game in of itself is typically not a venture backable company
2. If we knew the parent company and/or development company, games would be a more feasible investment opportuninity, but we are not given this information
3. Games are oftentimes on multiple platforms. Solely focusing on a game's mobile app performance will not holistically represent a game's true performance in the market
4. Games are nearly a third of all unique mobile apps in this dataset which will impact the dataset's overall statistics

```python
unique = unique[~unique.categoryPath.str.contains('game', case=False)]
```

```python
unique.describe()
```

### Cutoffs
1. Remove mobile apps with negative CMGR: Since we decided to use CMGR as the metric to predict an app's future performance, a negative CMGR indicates that growth has stalled and their user base has declined.
2. Focus on mobile apps with activeUsers in upper quartile: Mobile apps with low initial activeUsers skews CMGR since it's much easier to have a higher CMGR with initial numbers in the double and triple digits. 
3. Remove apps younger than six months: Similar to the reasoning behind cutoff \#2, mobile apps that have launched in the last couple months can have a much higher initial CMGR with very little historic data to prove their CMGR's sustainability. Mobile apps with an exceptionally high CMGR and inital activeUsers will be good to track for the next few months (more on this in the documentation). 

```python
# Remove mobile apps with negative CMGR
unique = unique[unique.cmgr > 0]

# Focus on mobile apps with activeUsers in upper quartile 
unique = unique[unique.activeUsers.between(unique.activeUsers.quantile(.75), unique.activeUsers.quantile(1.00))]

youngApps = unique[unique.startDate >= '2019-12-01']

# Remove apps younger than six months 
unique = unique[unique.startDate < '2019-12-01']
unique.plot(x='monthsActive', y='cmgr', kind='scatter')
```

# Top 15 mobile apps 

### Filter out for mobile apps with recentUsers in the top 35%
- I initially did not implement this hard cutoff because I thought that filtering startDate and activeUsers would inherently filter out mobile apps that had low current user numbers. However, I found that many of the startups had a lower than expected CMGR and recentUsers. 
- Having come to this conclusion, I initially fitlered for the top quartile of recentUsers, but this cutoff was too restrictive, leaving out many mobile apps on the fringe of this window, so I moved the lower quantile cutoff to 65%.

```python
final = unique[unique.recentUsers.between(unique.recentUsers.quantile(.65), unique.recentUsers.quantile(1.00))]
final.sort_values('cmgr', ascending=False).head(15)
```
